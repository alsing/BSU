# Least Squares Method

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return x ** 2 * np.sin(2 * x)
```

```python
x_values = [i for i in np.arange(-2, 2.1, 0.1)]
f_values = [f(i) for i in x_values]
```

```python
def find_approximation(x_vec, f_vec, n):
    s = [np.power(x_vec, i).sum() for i in range(2 * n + 1)]
    m = [np.power(x_vec, i).dot(f_vec).sum() for i in range(n + 1)]

    X = np.zeros((n + 1, n + 1))
    for i in range(n + 1):
        for j in range(n + 1):
            X[i][j] = s[j + i]

    y = m
    c = np.linalg.solve(X, y)

    q = [np.array([c[i] * x ** i for i in range(n + 1)]).sum() for x in x_values]
    return q
```

```python
q_values = find_approximation(x_values, f_values, 3)
delta = np.power(np.array(f_values) - np.array(q_values), 2).sum()
print('Delta (n = 3) = ', delta)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x_values, f_values, color='green', label='f(x)')
plt.scatter(x_values, q_values, color='red', label='$Q_n(x)$')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='best', prop={'size': 12})
plt.title('Least Squares Method (n=3)')
plt.show()
```
